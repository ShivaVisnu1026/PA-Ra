## Step 4: Install Project Dependencies

This project requires several Python packages. Let's install them from the requirements.txt file.


````python
# Install all required packages
import subprocess
import sys

def install_package(package):
    """Install a package using pip"""
    try:
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])
        print(f"✓ Successfully installed {package}")
        return True
    except subprocess.CalledProcessError:
        print(f"✗ Failed to install {package}")
        return False

# Read requirements.txt and install packages
try:
    with open('requirements.txt', 'r') as f:
        packages = [line.strip() for line in f if line.strip() and not line.startswith('#')]
    
    print(f"Found {len(packages)} packages to install:\n")
    for pkg in packages:
        print(f"  - {pkg}")
    
    print("\n" + "="*50)
    print("Installing packages...")
    print("="*50 + "\n")
    
    success_count = 0
    for pkg in packages:
        if install_package(pkg):
            success_count += 1
    
    print("\n" + "="*50)
    print(f"Installation complete: {success_count}/{len(packages)} packages installed")
    print("="*50)
    
except FileNotFoundError:
    print("⚠ requirements.txt not found. Installing packages manually...")
    packages = ['pandas', 'numpy', 'yfinance', 'streamlit', 'matplotlib']
    for pkg in packages:
        install_package(pkg)
````

## Step 5: Verify Package Installation

Let's verify that all required packages are installed correctly.


````python
# Verify all packages are installed
required_packages = {
    'pandas': 'pd',
    'numpy': 'np',
    'yfinance': 'yf',
    'streamlit': 'st',
    'matplotlib': 'plt'
}

print("Checking installed packages:\n")
print("-" * 50)

all_installed = True
for package, alias in required_packages.items():
    try:
        module = __import__(package)
        version = getattr(module, '__version__', 'unknown')
        print(f"✓ {package:15s} (version: {version})")
    except ImportError:
        print(f"✗ {package:15s} - NOT INSTALLED")
        all_installed = False

print("-" * 50)

if all_installed:
    print("\n✅ All packages are installed successfully!")
else:
    print("\n⚠ Some packages are missing. Please run the installation cell above.")
````

## Troubleshooting

### If Python is not found:
- Make sure Python is added to your PATH environment variable
- Restart your terminal/command prompt after installation
- Try using `py` instead of `python` on Windows

### If packages fail to install:
- Make sure you have internet connection
- Try upgrading pip first: `python -m pip install --upgrade pip`
- Some packages may require Visual C++ Build Tools on Windows

### If you get permission errors:
- Try using `--user` flag: `pip install --user package_name`
- Or use a virtual environment (recommended)

## Creating a Virtual Environment (Recommended)

Virtual environments help isolate project dependencies:

```bash
# Create virtual environment
python -m venv venv

# Activate on Windows
```
